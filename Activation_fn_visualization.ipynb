```python
import numpy as np
```

```python
# Sigmoid Function
# Formula: σ(x) = 1 / (1 + e^(-x))
def sigmoid(x):
    x = np.array(x)
    return 1 / (1 + np.exp(-x))
# Test Values
sig_input = [-8, -3, -1, 0.5, 3, 7]
sig_output = sigmoid(sig_input)

print("Sigmoid Input :", sig_input)
print("Sigmoid Output:", sig_output)
```

```python
# Tanh Function
# Formula: tanh(x) = (e^x − e^(-x)) / (e^x + e^(-x))

def tanh_act(x):
    return np.tanh(x)

tanh_input =  [-8, -3, -1, 0.5, 3, 7]
tanh_output = tanh_act(tanh_input)

print("Tanh Input :", tanh_input)
print("Tanh Output:", tanh_output)
```

```python
# ReLU Function
# Formula: ReLU(x) = max(0, x)

def relu(x):
    return np.maximum(0, x)

relu_input =  [-8, -3, -1, 0.5, 3, 7]
relu_output = relu(relu_input)

print("ReLU Input :", relu_input)
print("ReLU Output:", relu_output)
```

```python
# Leaky ReLU
# Formula: x if x > 0 else alphax

def leaky_relu(x, alpha=0.01):
    x = np.array(x)
    return np.where(x > 0, x, alpha * x)

leaky_input = [-8, -3, -1, 0.5, 3, 7]
leaky_output = leaky_relu(leaky_input)

print("Leaky ReLU Input :", leaky_input)
print("Leaky ReLU Output:", leaky_output)
```

```python
# ELU Function
# Formula: x if x > 0 else α(e^x − 1)

def elu(x, alpha=1.0):
    x = np.array(x)
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

elu_input =  [-8, -3, -1, 0.5, 3, 7]
elu_output = elu(elu_input)

print("ELU Input :", elu_input)
print("ELU Output:", elu_output)
```

```python
# Softmax Function
# Formula: e^(x_i) / Σ e^(x_j)

def softmax(z):
    z = np.array(z)
    exp_vals = np.exp(z - np.max(z))
    return exp_vals / np.sum(exp_vals)

soft_input =  [-8, -3, -1, 0.5, 3, 7]
soft_output = softmax(soft_input)

print("Softmax Input :", soft_input)
print("Softmax Output:", soft_output)
```

```python
inp = [[-8, -3, -1, 0.5, 3, 7]]

print("Input:", inp)
print("\n---------------------------")

print("\nSigmoid:")
print(sigmoid(inp))

print("\n---------------------------")

print("\nTanh:")
print(tanh_act(inp))

print("\n---------------------------")

print("\nReLU:")
print(relu(inp))

print("\n---------------------------")

print("\nLeaky ReLU:")
print(leaky_relu(inp))

print("\n---------------------------")

print("\nELU:")
print(elu(inp))

print("\n---------------------------")

print("\nSoftmax:")
print(softmax(inp))

print("\n---------------------------")
```

```python
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Input range
x = np.linspace(-10, 10, 200)
y = sigmoid(x)

# Plot
plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.title("Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
def tanh_act(x):
    return np.tanh(x)

# Input range
x = np.linspace(-8, 8, 200)
y = tanh_act(x)

# Plot
plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.title("Tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
# ReLU
def relu(x):
    return np.maximum(0, x)

# Input range
x = np.linspace(-5, 5, 200)
y = relu(x)

# Plot
plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
# Softmax
def softmax(x):
    exp_vals = np.exp(x - np.max(x))
    return exp_vals / np.sum(exp_vals)

# Input vector
x = np.array([-4, -1, 0.5, 2, 3.5, 6])
y = softmax(x)

# Bar Plot
plt.figure(figsize=(6,4))
plt.bar(range(len(x)), y)
plt.title("Softmax Activation Function")
plt.xlabel("Index")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```
