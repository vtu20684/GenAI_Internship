```python
# Assignment: Implementation of Loss Functions
import numpy as np
# True and Predicted Values
y_true = np.array([1, 0, 1, 1, 0])
y_pred = np.array([0.9, 0.2, 0.6, 0.4, 0.1])

print("True Values:", y_true)
print("Predicted Values:", y_pred)

# Mean Absolute Error (MAE)
# Formula: MAE = (1/n) * Σ |y_true - y_pred|
mae = np.mean(np.abs(y_true - y_pred))
print("\nMean Absolute Error (MAE):", mae)
```

```python
# Mean Squared Error (MSE)
# Formula: MSE = (1/n) * Σ (y_true - y_pred)^2
mse = np.mean((y_true - y_pred)**2)
print("Mean Squared Error (MSE):", mse)
```

```python
# Root Mean Squared Error (RMSE)
# Formula: RMSE = √MSE
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
# Binary Cross Entropy Loss
# Formula: BCE = - (1/n) * Σ [y_true log(y_pred) + (1-y_true) log(1-y_pred)]
bce = -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
print("\nBinary Cross Entropy Loss:", bce)
```

```python
# Categorical Cross Entropy (for multi-class example)
y_true = np.array([[1,0,0], [0,1,0], [0,0,1]])
y_pred = np.array([[0.8,0.1,0.1], [0.2,0.6,0.2], [0.1,0.2,0.7]])

cce = -np.mean(np.sum(y_true * np.log(y_pred), axis=1))
print("\nCategorical Cross Entropy Loss:", cce)
```
